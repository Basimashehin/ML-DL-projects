```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/dataset/Iris.csv')
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Species']=le.fit_transform(df['Species'])
```

```python
X=df.iloc[:,1:-1]
X
```

```python
Y=df.iloc[:,-1]
Y
```

```python
# Y
# 0->[1 0 0]
# 1->[0 1 0]
# 2->[0 0 1]
```

```python
from tensorflow.keras.utils import to_categorical
Y_new=to_categorical(Y)
Y_new
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
X_scaled

```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_scaled,Y_new,test_size=0.3,random_state=1)
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python

ann=Sequential()
# add an input layer
ann.add(Dense(units=4,activation='relu'))
# add hidden layer 1
ann.add(Dense(units=6,activation='relu'))
# add hidden layer 2
ann.add(Dense(units=10,activation='relu'))
# add output layer
ann.add(Dense(units=3,activation='softmax'))
```

```python
ann.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(X_train,Y_train,batch_size=10,epochs=100)
```

```python
ann.summary()
```

```python
loss,accuracy=ann.evaluate(X_test,Y_test)
print("loss:",loss)
print("accuracy:",accuracy)
```

```python
Y_pred=ann.predict(X_test)
Y_pred
```

```python
Y_pred1=Y_pred.argmax(axis=1)
```
