```python
import pandas as pd
from skimage.io import imread
from skimage.transform import resize
from keras.models import Sequential
from keras.layers import Dense,Conv2D,MaxPool2D,Flatten
import os
```

```python
os.listdir('/content/drive/MyDrive/Brain_Tumor/Testing')
```

```python
os.listdir('/content/drive/MyDrive/Brain_Tumor/Training')
```

```python
datadir='/content/drive/MyDrive/Brain_Tumor/Training'
```

```python
X_train=[]
Y_train=[]
categories=['pituitary', 'notumor', 'meningioma', 'glioma']
for i in categories:
  print("loading category.................",i)
  path=os.path.join(datadir,i)
  # print(path)
  images=os.listdir(path)
  # print(images)
  for img in images:
    imgpath=os.path.join(path,img)
    # print(imgpath)
    img_array=imread(imgpath)
    img_resized=resize(img_array,(150,150,1))
    X_train.append(img_resized)
    Y_train.append(categories.index(i))
  print('Loaded category',i)
```

```python
X_test=[]
Y_test=[]
categories=['pituitary', 'notumor', 'meningioma', 'glioma']
for i in categories:
  print("loading category.................",i)
  path=os.path.join('/content/drive/MyDrive/Brain_Tumor/Testing',i)
  # print(path)
  images=os.listdir(path)
  # print(images)
  for img in images:
    imgpath=os.path.join(path,img)
    # print(imgpath)
    img_array=imread(imgpath)
    img_resized=resize(img_array,(150,150,1))
    X_test.append(img_resized)
    Y_test.append(categories.index(i))
  print('Loaded category',i)
```

```python
import numpy as np
X_train=np.array(X_train)
X_test=np.array(X_test)
Y_train=np.array(Y_train)
Y_test=np.array(Y_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
model=Sequential()
model.add(Conv2D(32,(3,3),input_shape=(150,150,1),activation='relu'))
model.add(MaxPool2D(2,2))
# model.add(Conv2D(28,(3,3),activation='relu'))
# model.add(MaxPool2D(2,2))
model.add(Flatten())
model.add(Dense(100,activation='relu'))
model.add(Dense(4,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,Y_train,batch_size=32,epochs=10)
```

```python
loss,accuracy=model.evaluate(X_test,Y_test)
print('Loss:',loss)
print('accuracy:',accuracy)
```
